## Step 7: Handle Missing Values

If any missing values are present, they need to be handled before further processing.

In this notebook:

- Numerical columns will be filled using their mean value.
- Categorical columns will be filled with the value **"Unknown"**.

If the dataset does not contain any missing values, this step will not modify the data.

```python
# Fill missing values in numerical columns with their mean

numeric_columns = df.select_dtypes(include=['number']).columns

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

```python
# Fill missing values in categorical columns with 'Unknown'

categorical_columns = df.select_dtypes(include=['object']).columns

df[categorical_columns] = df[categorical_columns].fillna("Unknown")
```

# Delta Lake MERGE Implementation

After cleaning the dataset using Pandas, the next step is to perform incremental data processing using Delta Lake.

In this section, we will:

- Load the cleaned dataset into a Spark DataFrame.
- Convert columns to appropriate data types.
- Rename column names for Spark compatibility.
- Create a Delta table.
- Create an incremental dataset.
- Perform the MERGE operation.

## Step 1: Load the Cleaned Dataset into Spark

The cleaned CSV file generated during the Pandas section is loaded into a Spark DataFrame.

```python
spark_df = spark.read.csv(
    "/Volumes/workspace/default/assignment_data/cleaned_superstore.csv",
    header=True
)

display(spark_df)
```

## Step 2: Display the Schema

The schema is displayed to understand the structure of the dataset before performing any transformations.

```python
spark_df.printSchema()
```

## Step 3: Rename Column Names

Some column names contain spaces and hyphens.

To improve compatibility with Spark SQL and Delta Lake, spaces and hyphens are replaced with underscores.

```python
new_columns = [
    c.replace(" ", "_").replace("-", "_")
    for c in spark_df.columns
]

spark_df = spark_df.toDF(*new_columns)
```

```python
print(spark_df.columns)
```

## Step 4: Display the Updated DataFrame

The renamed DataFrame is displayed to verify that all column names have been updated successfully.

```python
display(spark_df)
```

## Step 5: Remove Duplicate Records

Duplicate records are removed to improve data quality before creating the Delta table.

```python
spark_df = spark_df.dropDuplicates()
```

```python
print("Total Records:", spark_df.count())
```

## Step 6: Create the Master Delta Table

The cleaned Spark DataFrame is stored as a Delta table named **customer_master**.

This table will be used as the target table during the Delta Lake MERGE operation.

```python
spark.sql("DROP TABLE IF EXISTS customer_master")
```

```python
spark_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("customer_master")
```

```python
display(spark.table("customer_master"))
```

```python
spark.table("customer_master").printSchema()
```

## Step 7: Load the Master Delta Table

The master Delta table is loaded into a Spark DataFrame.

This DataFrame will be used to generate an incremental dataset containing updated and new records.

```python
master_df = spark.table("customer_master")

display(master_df)
```

## Step 8: Load the Master Delta Table

The master Delta table is loaded into a Spark DataFrame.

This table will be used to create an incremental dataset containing both updated and new records.

```python
master_df = spark.table("customer_master")

display(master_df.limit(5))
```

## Step 9: Create Updated Records

To simulate an update, five existing records are selected from the master table.

The Sales and Profit values are increased to represent updated business data.

```python
from pyspark.sql.functions import col

updated_records = (
    master_df.limit(5)
    .withColumn("Sales", col("Sales") + 100)
    .withColumn("Profit", col("Profit") + 50)
)
```

```python
display(updated_records)
```

## Step 10: Create New Records

Five additional records are selected from the master table.

Their Row_ID values are modified to simulate completely new records that do not already exist in the master table.

```python
new_records = (
    master_df.limit(5)
    .withColumn("Row_ID", col("Row_ID") + 100000)
)
```

```python
display(new_records)
```

## Step 11: Create the Incremental Dataset

The updated records and new records are combined into a single incremental dataset.

This dataset represents newly arrived data that will be merged into the master Delta table.

```python
incremental_df = updated_records.unionByName(new_records)
```

```python
display(incremental_df)
```

## Step 12: Save the Incremental Dataset as a Delta Table

The incremental dataset is stored as a Delta table named **customer_incremental**.

This table will act as the source table during the Delta MERGE operation.

```python
spark.sql("DROP TABLE IF EXISTS customer_incremental")
```

```python
incremental_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("customer_incremental")
```

```python
display(spark.table("customer_incremental"))
```

```python
print("Incremental Records:", spark.table("customer_incremental").count())
```

## Step 13: Perform the Delta Lake MERGE Operation

The MERGE operation combines data from the incremental dataset into the master Delta table.

- Existing records (matched by `Row_ID`) are updated.
- New records (not matched) are inserted.

This demonstrates incremental data processing using Delta Lake.

```python
from delta.tables import DeltaTable

target_table = DeltaTable.forName(spark, "customer_master")
source_df = spark.table("customer_incremental")
```

```python
(
    target_table.alias("target")
    .merge(
        source_df.alias("source"),
        "target.Row_ID = source.Row_ID"
    )
    .whenMatchedUpdateAll()
    .whenNotMatchedInsertAll()
    .execute()
)
```

## Step 14: Verify the MERGE Operation

After the MERGE operation, the master table is checked to ensure that:

- Existing records have been updated.
- New records have been inserted successfully.

```python
display(spark.table("customer_master"))
```

```python
spark.sql("""
SELECT
    COUNT(*) AS total_records,
    COUNT(DISTINCT Row_ID) AS unique_row_ids
FROM customer_master
""").show()
```

## Step 15: Validate the Total Number of Records

After the MERGE operation, the total number of records is verified to ensure that new records have been inserted successfully while existing records have been updated.

```python
final_count = spark.table("customer_master").count()

print("Final Record Count:", final_count)
```

## Step 16: Validate Duplicate Records

The Row_ID column is checked to ensure that there are no duplicate records after performing the MERGE operation.

```python
spark.sql("""
SELECT
    Row_ID,
    COUNT(*) AS duplicate_count
FROM customer_master
GROUP BY Row_ID
HAVING COUNT(*) > 1
""").show()
```

## Step 17: Display Summary Statistics

Summary statistics are generated for the numerical columns to verify that the dataset has been successfully processed after the MERGE operation.

```python
spark.table("customer_master").describe(
    "Sales",
    "Quantity",
    "Profit",
    "Price",
    "total_amount"
).show()
```

## Step 18: Verify Updated Records

The updated records are displayed to confirm that the MERGE operation successfully updated the existing records in the master table.

```python
spark.sql("""
SELECT
    Row_ID,
    Sales,
    Profit
FROM customer_master
ORDER BY Row_ID
LIMIT 5
""").show()
```

## Step 19: Verify Newly Inserted Records

The newly inserted records are displayed by filtering records with Row_ID greater than 100000.

```python
spark.sql("""
SELECT
    Row_ID,
    Customer_Name,
    Sales,
    Profit
FROM customer_master
WHERE Row_ID > 100000
""").show()
```
